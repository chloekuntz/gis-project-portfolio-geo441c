## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)

- A geodatabase is the largest container for the feature datasets and the feature classes.
- The feature dataset is the next container, that is only smaller than the geodatabase.
- The feature datasets organize data based on relationships, such as a common spatial extent.
-The feature dataset provides a specification to which all the feature classes must adhere to; whether it's a boundary or something else.
- The feature classes are the smallest element, which reside within the feature datasets.
- The feature classes contain fundamental data, such as collections of all the lines or all the points.
